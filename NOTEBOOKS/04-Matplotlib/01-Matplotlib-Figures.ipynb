## Создаём объект Figure

Основная идея - создавать объекты и обращаться к их атрибутам и методам. Такой подход более удобен в случаях, когда на одном рисунке должны находиться несколько графиков. 

```python
# Создаём пустое поле для рисования (canvas/figure)
fig = plt.figure()
```

**ЗАМЕЧАНИЕ: ВСЕ КОМАНДЫ ДОЛЖНЫ НАХОДИТЬСЯ В ОДНОЙ ЯЧЕЙКЕ!**

Сначала мы создаём экземпляр figure. Далее мы добавляем оси (axes):

```python
# Создаём пустое поле для рисования (canvas/figure)
fig = plt.figure()

# Добавляем набор осей
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (диапазон от 0 до 1)

# Наносим график на этот набор осей
axes.plot(x, y)

plt.show()
```

### Добавляем ещё один набор осей

Для одного объекта figure можно добавить не только один набор осей, но и несколько таких наборов осей - с другими размерами и расположением. И далее мы можем рисовать графики отдельно на каждом из наборов осей.

```python
type(fig)
```

Код становится более сложным, но зато мы получаем полный контроль над тем, где располагаются те или иные графики. Обратите внимание, что в этом примере мы нарисуем график a,b два раза.

```python
# Создаём пустое поле для рисования (canvas/figure)
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])         # Большой набор осей
axes2 = fig.add_axes([0.2, 0.2, 0.5, 0.5]) # Маленький набор осей

# Рисуем график на бОльшем наборе осей - Axes 1
axes1.plot(a, b)

# С помощью методов set_ указываем атрибуты для осей
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Рисуем график на меньшем наборе осей - Axes 2
axes2.plot(a,b)
axes2.set_title('Small Figure');
```

Мы можем подвинуть меньший набор осей и поменять его параметры.

```python
# Создаём пустое поле для рисования (canvas/figure)
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])           # Большой набор осей
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # Маленький набор осей

# Рисуем график на бОльшем наборе осей - Axes 1
axes1.plot(a, b)

# С помощью методов set_ указываем атрибуты для осей
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Рисуем график на меньшем наборе осей - Axes 2
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed In');
```

Вы можете добавить несколько наборов осей, даже за пределами основного набора осей.

```python
# Создаём пустое поле для рисования (canvas/figure)
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])           # Большой набор осей
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # Маленький набор осей
axes3 = fig.add_axes([1, 1, 0.25, 0.25])     # Набор осей в верхнем правом углу!

# Рисуем график на бОльшем наборе осей - Axes 1
axes1.plot(a, b)

# С помощью методов set_ указываем атрибуты для осей
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Рисуем график на меньшем наборе осей - Axes 2
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed In');

# Рисуем график на третьем наборе осей - Axes 3
axes3.plot(a,b)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
np.random.seed(42)
x = np.linspace(0, 11, 10)
axes[0].plot(x, x**2 + 2 * x, label = 'func1', color = 'black', linewidth = 6, linestyle = '--', marker = 's', markersize = 12)
axes[0].plot(x, np.sqrt(x), label = "func2", color = 'purple', linewidth = 6)
fig.suptitle("Graphics")
axes[0].legend()

x = np.arange(-10, 10)
axes[1].plot(x, x**3)

fig.savefig("cubic and polimomial functions.png", bbox_inches  = 'tight')
```
